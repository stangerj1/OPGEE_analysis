```python
import pandas as pd
```

```python
df_in = pd.read_csv('./data/upstream_weekly-field_annual_production_kbd_2015-2023.csv')
list(df_in.columns)
```

```python
df = df_in[['country_name','field_name','year','metric_value','unit']]
df = df[df['country_name']=='Azerbaijan']

x=1
if all(df['unit']=='kboe/d'):
    print('All values are in kboe/d \n Multiplying by 365...')
    x=365
```

```python
result = df.groupby('year')['metric_value'].sum() * x
out = result.reset_index()
out
```

```python
out.to_csv('./data/AZE_prod.csv',index=False)
```
